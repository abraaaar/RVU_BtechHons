```python
# Introduce students to using RNNs for time series prediction by training an RNN to predict future values in a sine wave.
# Create a simple sine wave dataset.
# Train the RNN to predict the next value in the sine wave.
# Compare the predicted values with the actual sine wave.
import numpy as np
import matplotlib.pyplot as plt

# Generate sine wave data
timesteps = np.linspace(0, 10, 100)
data = np.sin(timesteps)
input_size = 1
hidden_size = 5
output_size = 1

W_xh = np.random.randn(hidden_size, input_size) * 0.01
W_hh = np.random.randn(hidden_size, hidden_size) * 0.01
W_hy = np.random.randn(output_size, hidden_size) * 0.01

def rnn_step(x, h_prev):
    h = np.tanh(np.dot(W_xh, x) + np.dot(W_hh, h_prev))
    y = np.dot(W_hy, h)
    return h, y

h = np.zeros((hidden_size, 1))
predictions = []

# Process the sine wave
for t in range(len(data) - 1):
    x = np.array([[data[t]]])
    h, y = rnn_step(x, h)
    predictions.append(y[0][0])

plt.plot(timesteps[:-1], data[:-1], label="Actual")
plt.plot(timesteps[:-1], predictions, label="Predicted")
plt.legend()
plt.show() 
# This activity demonstrates how RNNs can be used to predict future values in a time series, using a sine wave as the example. Students learn how RNNs can capture patterns in sequential data.
```
